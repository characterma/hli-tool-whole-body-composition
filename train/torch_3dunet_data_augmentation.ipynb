```
import numpy as np
import nibabel as nib
from loguru import logger
import warnings
warnings.filterwarnings("ignore")


class ImageEnhancement:
    def __init__(self):
        pass
    
    def image_rotate(self, image: np.array, label: np.array):
        """
        Rotate the image by a specified number of degrees
        将影像旋转指定度数， 因为当前影响不是立方体，为保持尺寸相同只能旋转180度
        :param image: origin image data, type: np.array
        :param label: origin label data, type: np.array
        """
        return [(np.rot90(image, 2, axes=(1, 2)), np.rot90(label, 2, axes=(1, 2)))]

    def horizontal_mirror(self, image: np.array, label: np.array):
        """
        The horizontal mirror image is inverted after 180 degrees of counterclockwise rotation
        水平镜像是将原始数据逆时针旋转180度，并进行逆转置 
        垂直镜像使用这行代码也可以： return [(np.flip(image, (1)), np.flip(label, (1)))]
        """
        return [(np.flip(image, (2)), np.flip(label, (2)))]
    
    def vertically_mirror(self, image: np.array, label: np.array):
        """
        The horizontal mirror image is inverted after 180 degrees of counterclockwise rotation
        水平镜像是将原始数据逆时针旋转180度，并进行逆转置
        """
        return [(np.rot90(image, 2)[::-1], np.rot90(label, 2)[::-1])]
    
    def image_mirror(self, image: np.array, label: np.array):
        h_mirror = self.horizontal_mirror(image=image, label=label)
        v_mirror = self.vertically_mirror(image=image, label=label)
        return h_mirror + v_mirror
    
#     def image_clip(self, image: np.array, label: np.array, slice_interval: list, n_range:int=20):
#         top_image, top_label = np.zeros_like(image), np.zeros_like(label)
#         behind_image, behind_label = np.zeros_like(image), np.zeros_like(label)
#         for i in range(image.shape[0]):
#             if i+20 < image.shape[0]:
#                 behind_image[i, :, :], behind_label[i:, :, :] = image[i+20, :, :], image[i+20, :, :]
                
#             if i-20 >= 0:
#                 top_image[i, :, :], top_label[i:, :, :] = image[i-20, :, :], image[i-20, :, :]        
#         return [(top_image, top_label), (behind_image, behind_label)]
    
    def image_clip(self, image: np.array, label: np.array, n_range:int=20):
        top_image, top_label = np.zeros_like(image), np.zeros_like(label)
        behind_image, behind_label = np.zeros_like(image), np.zeros_like(label)
        top_image[:image.shape[0]-20, :, :] = image[20:, :, :]
        top_label[:label.shape[0]-20, :, :] = label[20:, :, :]
        
        behind_image[20:, :, :] = image[:image.shape[0]-20, :, :]
        behind_label[20:, :, :] = label[:label.shape[0]-20:, :, :]
        return [(top_image, top_label), (behind_image, behind_label)]
    
    def dimension_conversion(self, arr):
        imgs = np.rot90(arr, 3)
        transform_dicom_array = np.zeros((556, 260, 320))
        for i, j in zip(range(imgs.shape[-1]), range(imgs.shape[-1]-1, 0-1, -1)):
            transform_dicom_array[i, :, :] = imgs[:, :, j]
        return transform_dicom_array
        
    
    def image_fat_include_noise(self, image_path: str, vat_path: str, asat_path: str):
        image = nib.load(image_path).get_data() 
        vat = nib.load(vat_path).get_data()
        asat = nib.load(asat_path).get_data()
        label_list = [vat, asat]
        # todo: merge label
        if label_list.__len__() > 0:
            img_bg = np.zeros_like(label_list[0], dtype=int)
        for index, label in enumerate(label_list):
            process_label = np.where(label==1, index+1, 0)
            img_bg = img_bg + process_label
        logger.info(f"image shape: {image.shape}, label shape: {img_bg.shape}")
        image = self.dimension_conversion(image)
        img_bg = self.dimension_conversion(img_bg)
        logger.info(f"image shape: {image.shape}, label shape: {img_bg.shape}")

        return [(image, img_bg)]
    
    def image_water_include_noise(self, image_path: str, muscle_path: str):
        image = nib.load(image_path).get_data()
        muscle = nib.load(muscle_path).get_data()
        logger.info(f"image shape: {image.shape}, label shape: {muscle.shape}")
        image = self.dimension_conversion(image)
        muscle = self.dimension_conversion(muscle)
        logger.info(f"image shape: {image.shape}, label shape: {muscle.shape}")
        return [(image, muscle)]
    
    def excute_fat(self, image: np.array, label: np.array, image_path:str, vat_path:str, asat_path: str):
        rot_image = self.image_rotate(image, label)
        print(image.shape, label.shape)
        mirror_image = self.image_mirror(image, label)
        clip_image = self.image_clip(image, label)
        noise_image = self.image_fat_include_noise(image_path, vat_path, asat_path)
        return rot_image + mirror_image + clip_image + noise_image
    
    def excute_water(self, image: np.array, label: np.array, image_path:str, muscle_path:str):
        rot_image = self.image_rotate(image, label)
        mirror_image = self.image_mirror(image, label)
        clip_image = self.image_clip(image, label)
        noise_image = self.image_water_include_noise(image_path, muscle_path)
        return rot_image + mirror_image + clip_image + noise_image
        return mirror_image
```
