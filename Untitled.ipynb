```python
def create_overlays(predicted, dcm_objs):
    processed_predictions = []
    for i in range(0, len(predicted)):
        # Assign each pixel to highest predicted probability

        original = dcm_objs[i].pixel_array
        cvimage = np.floor((original / original.max()) * 255) 
        cvimage = cvimage.astype('uint8')

        output = np.array(cvimage.copy())
        output = np.stack((np.squeeze(output),) * 3, -1)

        masks = np.zeros(
            (
                predicted[i].shape[0],
                predicted[i].shape[1],
                3
            ),
            dtype='uint8'
        )
        colors = {
                0: [206, 68, 100],
                1: [68, 77, 206]
            }
    
        for key, value in colors.items():
            masks[predicted[i] == key] = value

        masks = gaussian(
            masks,
            sigma=.9,
            mode='wrap',
            multichannel=True,
            preserve_range=True
        ).astype('uint8')

        cv2.addWeighted(masks, 0.3, output, 0.7, 0, masks)
        processed_predictions.append(masks)  # do we apply masks over images?
    return processed_predictions
```

```python
dicom_path = "./scratch/working_local/images/"
path = "./scratch/working_local/images/"
imgs = [dcm.read_file(os.path.join(dicom_path, path)) for path in os.listdir(dicom_path)]
```
